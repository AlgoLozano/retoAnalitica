#### Describe nos ayudará a observar rangos y ver inicialmente cómo se comportan los datos, ahora con base en esto se pueden graficar algunos datos para observar su correlación

```python
import matplotlib.pyplot as plt
import numpy as np
 
col1 = datos['Total Volume']
col2 = datos['4046']
col3 = datos['4225']
col4 = datos['4770']

grafica =[col1,col2,col3,col4]

fig = plt.figure(figsize =(10, 7))

plt.boxplot(grafica, notch ='True')
plt.ylabel("Cantidad")
plt.xlabel("Volumen total                             4046                                   4225                               4770")
plt.title("Relación entre calibre de aguacate y volumen total")

plt.show()
```

#### La gráfica anterior muestra hasta la izquierda el volumen total de aguacates y las siguientes tres columnas son referentes a cada tipo de aguacate (refiriéndome al calibre por el PLU). Se puede observar cómo los tres tipos forman el volumen total. Con esto podemos inferir que el volumen se refiere a la cantidad de aguacates en sí.

```python
import matplotlib.pyplot as plt
import numpy as np
 
col1 = datos['Total Bags']
col2 = datos['4046']
col3 = datos['4225']
col4 = datos['4770']

grafica =[col1,col2,col3,col4]

fig = plt.figure(figsize =(10, 7))

plt.boxplot(grafica, notch ='True')
plt.ylabel("Cantidad")
plt.xlabel("Bolsas totales                             4046                                   4225                               4770")
plt.title("Relación entre cantidad de bolsas y calibre de aguacate")

plt.show()
```

#### La gráfica anterior es esencial ya que descarta la idea de una relación entre las bolsas y cada tipo de aguacate a menos de que exista una cantidad de 4 a 5 aguacates 4046 por bolsa y es por ello que ahora se puede obtener información para comparar las bolsas totales y los tipos de bolsa.

```python
import matplotlib.pyplot as plt
import numpy as np
 
col1 = datos['Total Bags']
col2 = datos['Small Bags']
col3 = datos['Large Bags']
col4 = datos['XLarge Bags']

grafica =[col1,col2,col3,col4]

fig = plt.figure(figsize =(12, 7))

plt.boxplot(grafica, notch ='True')
plt.ylabel("Cantidad")
plt.xlabel("Bolsas totales                             Bolsas pequeñas                                   Bolsas grandes                               Bolsas extra grandes")
plt.title("Relación entre bolsas totales y cada tipo de bolsa")

plt.show()
```

#### Lo anterior implica que hay una cantidad mínima de bolsas extra grandes, y la mayoría es de bolsas pequeñas, esto quiere decir que es más conveniente exportar y vender por bolsas pequeñas que por bolsas grandes o extra grandes.
